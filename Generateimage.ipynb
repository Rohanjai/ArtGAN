```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
generator = torch.load('/Users/rohanjai/Developer/ArtGAN/model-weights/generator.pth',map_location=('cpu'))
```

```python
stats = (0.5, 0.5, 0.5), (0.5, 0.5, 0.5)
latent_dim = 16
```

```python
def denorm(img_tensors):
    return img_tensors * stats[1][0] + stats[0][0]
```

```python
def print_image(image):
    fig, ax = plt.subplots(figsize=(3, 3))
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(denorm(image))
    plt.show()
```

```python
fixed_latent = torch.randn(latent_dim, latent_dim, 1, 1, device='cpu')
fake_images = generator(fixed_latent)
print_image(fake_images.detach().cpu().numpy()[0].transpose(1,2,0))
```
